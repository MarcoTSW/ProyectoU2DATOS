El objetivo de esto es poder implementar un grafo usando una lista de adyacencia y asi aplicar un algoritmo greedy para la coloración de vertices.

Conceptos clave:

- Grafo: Es una estructura de nodos(vertices) y conexiones(aristas)
- Lista de adyacencia: Es la representación mediante diccionario
- Algoritmo greedy: Toma las decisiones locales óptimas

Se implementó una clase Graph que permite:

- Agregar vértices y aristas
- Obtener vecinos
- Aplicar coloración greedy
- Validar la coloración

```python
class Graph:
    def __init__(self):
        self.adjacency = {}

    def add_vertex(self, v):
        if v not in self.adjacency:
            self.adjacency[v] = []

    def add_edge(self, u, v):
        self.add_vertex(u)
        self.add_vertex(v)
        if v not in self.adjacency[u]:
            self.adjacency[u].append(v)
        if u not in self.adjacency[v]:
            self.adjacency[v].append(u)

    def neighbors(self, v):
        return self.adjacency.get(v, [])

    def vertices(self):
        return list(self.adjacency.keys())

    def greedy_coloring(self, order=None):
        if order is None:
            order = sorted(self.vertices())

        coloring = {}

        for v in order:
            neighbor_colors = set(
                coloring[n] for n in self.neighbors(v) if n in coloring
            )

            color = 0
            while color in neighbor_colors:
                color += 1

            coloring[v] = color

        return coloring

    def greedy_coloring_by_degree(self):
        order = sorted(self.vertices(), key=lambda v: len(self.adjacency[v]), reverse=True)
        return self.greedy_coloring(order=order)

    def validate_coloring(self, coloring):
        for v in self.vertices():
            for n in self.neighbors(v):
                if coloring.get(v) == coloring.get(n):
                    return False
        return True
```

```python
def mostrar_resultados(coloring, valid):
    print("\n--- Resultados de coloración ---")
    for vertice, color in sorted(coloring.items()):
        print(f"Vértice {vertice} -> Color {color}")
    num_colors = len(set(coloring.values()))
    print(f"\nNúmero total de colores usados: {num_colors}")
    print(f"Validación: {'Válida ✓' if valid else 'No válida ✗'}")
```

## Prueba 1: Grafo completo K4
Se espera usar 4 colores porque todos los vértices están conectados.

```python
def test_k4():
    g = Graph()
    vertices = ["A", "B", "C", "D"]

    for v in vertices:
        g.add_vertex(v)

    for i in range(len(vertices)):
        for j in range(i + 1, len(vertices)):
            g.add_edge(vertices[i], vertices[j])

    print("\n=== PRUEBA K4 ===")
    coloring = g.greedy_coloring()
    mostrar_resultados(coloring, g.validate_coloring(coloring))

test_k4()
```

## Prueba 2: Grafo ciclo C5

En esta prueba se construye un grafo en forma de ciclo con 5 vértices (C5), donde cada vértice está conectado únicamente con dos vecinos formando un ciclo cerrado.

Al tratarse de un ciclo impar, no es posible colorearlo con solo 2 colores sin generar conflictos.

Se espera que el algoritmo greedy utilice 3 colores.

Esta prueba permite analizar el comportamiento del algoritmo en estructuras cíclicas y verificar su correcto funcionamiento.

```python
def test_c5():
    g = Graph()
    vertices = ["A", "B", "C", "D", "E"]

    for v in vertices:
        g.add_vertex(v)

    for i in range(len(vertices)):
        g.add_edge(vertices[i], vertices[(i + 1) % len(vertices)])

    print("\n=== PRUEBA C5 ===")
    coloring = g.greedy_coloring()
    mostrar_resultados(coloring, g.validate_coloring(coloring))

test_c5()
```

## Prueba 3: Comparación de órdenes

En esta prueba se utiliza un mismo grafo para aplicar el algoritmo greedy con dos diferentes órdenes de procesamiento de vértices:

- Orden natural (alfabético)
- Orden por grado descendente (vértices con más conexiones primero)

El objetivo es comparar cómo el orden de los vértices afecta el número total de colores utilizados.

Se espera observar que el algoritmo greedy no siempre produce la solución óptima, ya que depende del orden en que se procesan los vértices.

Esta prueba demuestra una de las principales características de los algoritmos greedy.

```python
def test_comparacion():
    g = Graph()

    for v in ["A", "B", "C", "D", "E"]:
        g.add_vertex(v)

    g.add_edge("A", "B")
    g.add_edge("A", "C")
    g.add_edge("B", "C")
    g.add_edge("B", "D")
    g.add_edge("C", "E")
    g.add_edge("D", "E")

    print("\n=== COMPARACIÓN DE ÓRDENES ===")

    print("\n--- Orden natural ---")
    col1 = g.greedy_coloring()
    mostrar_resultados(col1, g.validate_coloring(col1))

    print("\n--- Orden por grado descendente ---")
    col2 = g.greedy_coloring_by_degree()
    mostrar_resultados(col2, g.validate_coloring(col2))

test_comparacion()
```

## Conclusiones

El algoritmo greedy depende del orden de los vértices.
No siempre produce el número mínimo de colores, pero es eficiente.
